# Ligand-based Virtual Screening

```python
import pandas as pd
import numpy as np
import datamol as dm
from molfeat.trans.base import MoleculeTransformer
from molfeat.trans.fp import FPVecTransformer
from sklearn.metrics.pairwise import cosine_similarity
import matplotlib.pyplot as plt
from collections import defaultdict
import tqdm
from rdkit.Chem import FilterCatalog
from rdkit.Chem.FilterCatalog import FilterCatalogParams

def get_feature(method, smiles_list):
    if method == 'ecfp':
        transformer = FPVecTransformer(kind='ecfp', dtype=float)
        features = transformer(smiles_list)
    elif method == 'maccs':
        transformer = FPVecTransformer("maccs", dtype=np.float32)
        features = transformer(smiles_list)
    return features

def tanimoto_similarity_numpy(a, b):
    intersection = np.sum(a * b)
    union = np.sum(a) + np.sum(b) - intersection
    if union == 0:
        return 0.0
    return float(intersection) / union

def get_top_n_compounds(method, n, screen_features, test_df):
    # Calculate similarity between each screen_feature and all test_features
    test_features = get_feature(method, test_df['smiles'])

    max_similarities = []

    for test_feat in tqdm.tqdm(test_features):
        similarities = [tanimoto_similarity_numpy(screen_feat, test_feat) for screen_feat in screen_features]
        max_similarities.append(max(similarities))

    # 유사도 정보를 포함한 복사본 생성
    test_df = test_df.copy()
    test_df['max_similarity'] = max_similarities

    # max_similarity 기준으로 내림차순 정렬
    sorted_df = test_df.sort_values(by='max_similarity', ascending=False).reset_index()

    # 상위 N개 선택
    top_n_df = sorted_df.head(n)
    top_n_indices = top_n_df['index'].values
    top_n_scores = top_n_df['max_similarity'].values
    top_n_compounds = top_n_df.drop(columns=['index'])

    # positive ratio 계산
    positive_ratio = top_n_df['y'].sum() / n
    return top_n_df, top_n_indices, top_n_scores, positive_ratio, top_n_compounds

# Load data
screen_df = pd.read_csv('./PKM2/original/screen.csv')
test_df = pd.read_csv('./PKM2/original/test.csv')

# Filter positive labels only
positive_label_screen_df = screen_df[screen_df['y']==1]

# List of SMILES from screen data (using only 2 samples)
screen_smiles_list = positive_label_screen_df['smiles']

# Evaluate performance for various n values
n_fixed = 100  # Fixed n value for comparison

random_samples = test_df.sample(n=n_fixed, random_state=42)  # random_state는 재현성을 위해 설정
test_df = test_df
positive_count = sum(random_samples['y'] == 1)
pos_ratio = positive_count / n_fixed
print(f"Random Positive ratio = {pos_ratio:.4f}")
# Compare various feature extraction methods (additional functionality)
methods = ['ecfp', 'maccs']

print("\nComparison of various feature extraction methods (n=100):")

for method in methods:
    screen_features = get_feature(method, screen_smiles_list)
    top_n_df, top_n_indices, top_n_scores, pos_ratio, top_n_compounds = get_top_n_compounds(method, n_fixed, screen_features, test_df)
    print(f"{method}: Positive ratio = {pos_ratio:.4f}")
    top_n_df.to_csv('./Top_Hits_%s.csv'%(method))
```

# Clustring & Selection

```python
import pandas as pd
import numpy as np
import datamol as dm
from molfeat.trans.base import MoleculeTransformer
from molfeat.trans.fp import FPVecTransformer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.manifold import TSNE
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
from collections import defaultdict
import tqdm

def get_feature(method, smiles_list):
    if method == 'ecfp':
        transformer = FPVecTransformer(kind='ecfp', dtype=float)
        features = transformer(smiles_list)
    elif method == 'maccs':
        transformer = FPVecTransformer("maccs", dtype=np.float32)
        features = transformer(smiles_list)

    return features

def cluster_compounds(features, n_clusters=5):
    # Scale the features for better clustering performance
    scaler = StandardScaler()
    scaled_features = scaler.fit_transform(features)

    # Perform K-means clustering
    kmeans = KMeans(n_clusters=n_clusters, random_state=42, n_init=10)
    cluster_labels = kmeans.fit_predict(scaled_features)

    return cluster_labels, scaled_features

def visualize_with_tsne_colored_by_clusters(features, labels, perplexity=30, n_iter=1000):
    # Apply t-SNE for dimensionality reduction
    tsne = TSNE(n_components=2, perplexity=perplexity, n_iter=n_iter,
                random_state=42, learning_rate='auto')
    tsne_features = tsne.fit_transform(features)

    # Plot the t-SNE projection with clusters colored differently
    plt.figure(figsize=(12, 10))

    # Create a colormap for the clusters
    colors = plt.cm.tab10(np.linspace(0, 1, len(np.unique(labels))))

    # Plot each cluster with a different color
    for i, cluster_id in enumerate(np.unique(labels)):
        cluster_points = tsne_features[labels == cluster_id]
        plt.scatter(cluster_points[:, 0], cluster_points[:, 1],
                    color=colors[i], label=f'Cluster {cluster_id}',
                    alpha=0.7, s=70, edgecolors='w', linewidth=0.5)

    plt.title('t-SNE Projection of Chemical Compounds', fontsize=16)
    plt.xlabel('t-SNE1', fontsize=14)
    plt.ylabel('t-SNE2', fontsize=14)
    plt.legend(fontsize=12)
    plt.grid(True, linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()

    return tsne_features

# Main code
test_df = pd.read_csv('./PKM2/original/test.csv')
candidate_df = test_df[test_df['y']==1]
candidate_smiles_list = candidate_df['smiles']

# Get molecular features
method = 'maccs'
features = get_feature(method, candidate_smiles_list)

# Define number of clusters
n_clusters = 10  # You can change this to your desired number of clusters

# Perform clustering
cluster_labels, scaled_features = cluster_compounds(features, n_clusters=n_clusters)

# Add cluster labels to the candidate_df
candidate_df['cluster'] = cluster_labels

# Display cluster distribution
print(f"Cluster distribution:")
cluster_counts = pd.Series(cluster_labels).value_counts().sort_index()
for cluster_id, count in cluster_counts.items():
    percentage = count / len(cluster_labels) * 100
    print(f"Cluster {cluster_id}: {count} compounds ({percentage:.2f}%)")

# Visualize with t-SNE colored by clusters
print("Performing t-SNE dimensionality reduction. This may take a moment...")
tsne_features = visualize_with_tsne_colored_by_clusters(scaled_features, cluster_labels)

# Create a dataframe with t-SNE coordinates for further analysis (optional)
tsne_df = pd.DataFrame(tsne_features, columns=['TSNE1', 'TSNE2'])
tsne_df['cluster'] = cluster_labels
tsne_df['smiles'] = candidate_smiles_list.reset_index(drop=True)

# Save the updated dataframe with cluster information
candidate_df.to_csv('./candidates_with_clusters.csv', index=False)

print(f"Clustering completed. Added cluster information to candidate dataframe.")
```

# Visualization

```python
# 필요한 라이브러리 import
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from rdkit import Chem
from rdkit.Chem import Draw
from rdkit.Chem import AllChem
from rdkit.Chem.Draw import IPythonConsole
from IPython.display import display

# 클러스터별 대표 화합물 시각화 함수 정의
def visualize_compounds_by_cluster(df, n_per_cluster=5, molsPerRow=3, subImgSize=(200, 200)):
    # 고유한 클러스터 ID 가져오기
    unique_clusters = sorted(df['cluster'].unique())

    for cluster_id in unique_clusters:
        # 현재 클러스터의 화합물만 필터링
        cluster_compounds = df[df['cluster'] == cluster_id]

        # n_per_cluster개만 샘플링 (또는 전체 사용)
        if len(cluster_compounds) > n_per_cluster:
            sampled_compounds = cluster_compounds.sample(n_per_cluster, random_state=42)
        else:
            sampled_compounds = cluster_compounds

        # SMILES 문자열을 RDKit 분자 객체로 변환
        mols = []
        valid_smiles = []

        for idx, row in sampled_compounds.iterrows():
            mol = Chem.MolFromSmiles(row['smiles'])
            if mol is not None:  # 유효한 SMILES 문자열만 처리
                mols.append(mol)
                valid_smiles.append(row['smiles'])

        # 2D 구조 생성
        for mol in mols:
            AllChem.Compute2DCoords(mol)

        # 화합물 시각화
        if mols:
            print(f"클러스터 {cluster_id} 대표 화합물 ({len(mols)}개)")
            img = Draw.MolsToGridImage(mols, molsPerRow=molsPerRow, subImgSize=subImgSize,
                                     legends=[f"Compound {i+1}" for i in range(len(mols))])
            display(img)

            # SMILES 문자열 표시 (참조용)
            for i, smiles in enumerate(valid_smiles):
                print(f"Compound {i+1}: {smiles}")
            print("\n" + "-"*80 + "\n")
        else:
            print(f"클러스터 {cluster_id}에는 유효한 화합물이 없습니다.")

# candidates_with_clusters.csv 파일 로드
candidates_with_clusters = pd.read_csv('./candidates_with_clusters.csv')

# 데이터 확인
print(f"총 화합물 수: {len(candidates_with_clusters)}")
print(f"클러스터 분포:\n{candidates_with_clusters['cluster'].value_counts().sort_index()}")
candidates_with_clusters.head()

visualize_compounds_by_cluster(candidates_with_clusters, n_per_cluster=5)
```
